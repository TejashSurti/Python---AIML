### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
#dropping a few columns
dfmovie1 = dfmovie.drop(['movie id','sum','movie title','release date'], axis=1)

dfmovie_groupdata = dfmovie1.groupby(by=['release year']).sum()
dfmovie_groupdata
```

```python
#gettting the total number of releases in a year
dfmovie_groupdata['total release'] = dfmovie.groupby(by=['release year'])['movie title'].count()
dfmovie_groupdata
```

```python
#Getting the percentages and rounding to 2 decimals.
dfmovie_final = round(dfmovie_groupdata.div(dfmovie_groupdata['total release'], axis= 0) * 100,2)
dfmovie_final.head()
```

```python
#dropping total release column before plotting as it is no longer needed.
dfmovie_final.drop('total release',axis=1, inplace=True)
dfmovie_final.head()
```

```python
plt.figure(figsize=(20,20))
relase_popularity_plot = sns.heatmap(dfmovie_final, annot=True, linewidths=.5, fmt= '.1f', center = 1 )  # heatmap
relase_popularity_plot.tick_params(top=True, labeltop=True) #showing labels on top as well since plot is big.
plt.show()
```

**Insights:**
During the initial years only few specific genres were popular, partly also because there were fewer releases. 
As year's passed the popularity for different genres started spreading out. However, drama still remains the most popular genre followed by thriller and comedy at the end of 1998.


### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#apply group by and aggregating on count of ratings and avg rating. 
dftop100movies =dfrating.groupby(['movie id'])['rating'].agg([('rating count','count'),('rating avg','mean')]).reset_index()

#filtering to only keep movies that have more than 100 ratings
dftop100movies = dftop100movies.loc[dftop100movies['rating count']>100]

#sorting by avg rating
dftop100movies.sort_values(by=['rating avg'], ascending=False, inplace=True)
```

```python
#getting top 25 movies based on avg rating
dftop25movie = dftop100movies.head(25)
dftop25movie
```

```python
#finally merging the filtered dataframe with top 25 movies based on avg rating and the movie dataframe with details on the movie.
pd.merge(dftop25movie,dfmovie, how='inner',on='movie id')
```

```python
#merging all the three data sets. Using nested merges to firt get merged data set on two frames and then get the final data.
dffinaldata = pd.merge(pd.merge(dfrating,dfmovie, how='inner', on='movie id'),dfuser, how='inner', on='user id')
dffinaldata.head()
```

```python
dffinaldata = dffinaldata.drop(['user id','movie id','rating','timestamp','movie title','release date','sum',
                               'release year','age','occupation','zip code'], axis=1)
dffinaldata.head()
```

```python
dffinaldata = dffinaldata.groupby(by='gender').sum()
```

```python
dffinaldata['total ratings'] = dffinaldata.sum(axis=1)
dffinaldata.head()
```

```python
dffinaldata = round(dffinaldata.div(dffinaldata['total ratings'], axis= 0) * 100,2)
dffinaldata
```
